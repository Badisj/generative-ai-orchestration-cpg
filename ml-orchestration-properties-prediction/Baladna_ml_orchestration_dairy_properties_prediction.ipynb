### 3. Training Loop

```python
import subprocess

cmd = [
    "python",
    "./src/train.py",
    "--input", "./data/processed/Recipe_yogurt_data_generated_final_processed.csv",
    "--out-dir", "./models",
    "--cv-folds", "3",
    "--trials", "5"
]

# Run the process and stream output
process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)

# Print each line as it arrives
for line in iter(process.stdout.readline, ''):
    print(line, end='')  # end='' avoids double newlines

process.stdout.close()
process.wait()
```
